```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv("cleaned_data.csv")
```

```python
ds.info()
```

```python
ds.describe()
```

```python
#Companywise count of cars
companywise_count = ds["company"].value_counts()
# Plotting the bar chart
ax = companywise_count.plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')

plt.title("Number of Cars by Company")
plt.xlabel("Company")
plt.ylabel("Count of Cars")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with count values
for i, count in enumerate(companywise_count):
    ax.text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#yearwise count of cars
yearwise_count = ds["year"].value_counts()
yearwise_count.plot(kind = 'bar')
plt.show()
```

```python
#Fuel Type wise count
fuel_type_counts = ds["fuel_type"].value_counts()
fuel_type_counts.plot(kind = 'pie', autopct='%1.1f%%')
plt.show()
```
